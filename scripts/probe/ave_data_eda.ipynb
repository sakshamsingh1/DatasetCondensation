```python
path = os.path.join(base_path,'Annotations.txt')
df = pd.read_csv(path, delimiter='&')
df.head()
```

```python
df['len'] = df.apply(lambda x : x['EndTime']-x['StartTime'],axis=1)
df.head()
```

```python
# create a barplot of category grouped by length
# groupby category and sum of length
df_g = df.groupby('Category')['len'].sum().reset_index()

sns.set(style="whitegrid")
ax = sns.barplot(x="Category", y="len", data=df)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.show()
```
